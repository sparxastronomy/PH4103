```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("Sheet3.csv")
```

```python
df
```

```python
T = df["Temperature"][:]
R = df["Hall Coefficient"][:]
```

```python
plt.figure(figsize=(6, 4), dpi=100)
plt.axhline(0, color="black")
plt.plot(T, R, "--o", mfc='red', ms=8)

plt.grid(alpha=0.5)

plt.xlabel("Temperature ($^\\circ$C)", fontsize=12)
plt.ylabel("Hall Coefficient ($R$)", fontsize=12)

plt.savefig("Hall_Coefficient.jpg", dpi=300, bbox_inches="tight")
```

```python
# calulte the charge density
q = 1.60217662e-19
n = 1 / (R*q)
```

```python
# plot the charge density
plt.figure(figsize=(6, 4), dpi=100)
plt.axhline(0, color="black")
plt.plot(T[0:-3], n[0:-3], "--o", mfc='red', ms=8)

plt.grid(alpha=0.5)

plt.xlabel("Temperature ($^\\circ$C)", fontsize=12)
plt.ylabel("Charge Density ($n$)", fontsize=12)
plt.savefig("Charge_Density.jpg", dpi=300, bbox_inches="tight")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# change font to Times New Roman
plt.rcParams['font.family'] = 'Times New Roman'
```

```python
df = pd.read_csv('Dielectric.csv')
df.head(10)
```

```python
Temp = df['Temperature']
Cap  = df['Capacitance'] # in pF
```

```python
plt.plot(Temp[1:], Cap[1:], 'o')
```

```python
E0 =  8.85418782e-12 #permittivity of free space in F/m
A  =  48 #area of the plate in mm^2
d  =  1.42 #distance between the plates in mm

# calculate C0 as E0*A/d in SI units
C0 = E0*A/d*1e-3 

# calculate the capacitance in pF
C0 = C0*1e12
print('C0 =', C0, 'pF')

# effective capacitance
Ceff = Cap/C0
```

```python
import seaborn as sns
sns.set_style('ticks')
# increase the font size of tick labels
plt.rcParams['xtick.labelsize'] = 13
plt.rcParams['ytick.labelsize'] = 13


plt.figure(figsize=(6,7), dpi=100)
plt.plot(Temp[1:], Ceff[1:], '-o', mfc='orange', ms=8, label='Data points')

#add a vertical and horizontal line at the peak of the curve
plt.axvline(x=Temp[1:][np.argmax(Ceff[:])],ymax=0.95, color='k', linestyle='--', label='T$_C$ = %2.1f $^\\circ$C'%Temp[1:][np.argmax(Ceff[:])])
plt.axhline(y=Ceff[1:][np.argmax(Ceff[:])],xmax=0.74, color='k', linestyle='--', label='$\\epsilon$ = %2.1f '%Ceff[1:][np.argmax(Ceff[:])])


plt.grid(alpha=0.5)

plt.legend(loc='best', fontsize=13)
plt.xlabel('Temperature ($^\circ$C)', fontsize=14)
plt.ylabel('Dielectric Constant ($\\varepsilon$)', fontsize=14)

plt.savefig('Dielectric.jpg',dpi=300, bbox_inches='tight')
```
